## PRAW

You may use any method that you prefer to access Reddit and pull the data, here I am showing you a python library designed to do just this.

The documentation is here:  https://praw.readthedocs.io/en/latest/index.html

```python
import json
import base64
import requests
from requests.exceptions import ConnectionError
import getpass

from os import path
from cryptography.fernet import Fernet

# read your credentials back

def decrypt(encrypted_message):
    key = load_key()
    f = Fernet(key)
    decrypted_message = f.decrypt(encrypted_message)
    return decrypted_message.decode()

def load_key():
    return open("/home/jovyan/secret.key", "rb").read()

if (path.exists('/home/jovyan/reddit_encypt.json')):
    with open('/home/jovyan/reddit_encypt.json','rb') as f:
        bytes = f.read()
    creds = json.loads(decrypt(bytes))
else:
    print('Path not found')
    
myclientid = creds['client_id']
mysecret = creds['secret']
mypassword = creds['password']
myusername = creds['username']
```

```python
reddit = praw.Reddit(
    client_id=myclientid,
    client_secret=mysecret,
    password=mypassword,
    user_agent="wsb scraper by morningstar inc",
    username=myusername
)

print(reddit.user.me())


submission = reddit.submission(id="m8hqmj")
print (submission)
```

```python
for top_level_comment in submission.comments:
    if isinstance(top_level_comment, praw.models.reddit.comment.Comment):
        print(top_level_comment)
```
